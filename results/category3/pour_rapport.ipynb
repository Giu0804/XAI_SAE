
### Catégorie 3 : Preuves Causales et Interventions (Mechanistic Interpretability)

#### 1. Problématique et Hypothèse Fondamentale
Jusqu'à présent, les analyses de corrélations (Catégorie 1) ont démontré que la caractéristique SAE s'active en présence du concept d'ADN. Cependant, une critique majeure subsiste : l'autoencodeur ayant été entraîné uniquement sur les activations internes, ces caractéristiques pourraient n'être que de simples artefacts statistiques reflétant les corrélations des données d'entraînement, sans aucune pertinence pour le fonctionnement réel du réseau.

**L'hypothèse :** Si la caractéristique identifiée (ex: l'ADN) correspond véritablement au mécanisme interne par lequel le modèle "pense" à ce concept, alors intervenir directement sur cette caractéristique doit modifier le comportement en aval du modèle de manière prévisible. L'objectif de la Catégorie 3 est donc d'ouvrir la "boîte noire" et de prouver la causalité par des interventions chirurgicales sur les activations (techniques de *Hooks*).

#### 2. Expérience 3A : L'Ablation Causale (Feature Ablation)
**L'objectif :** Prouver que la caractéristique est indispensable au modèle pour traiter un concept précis. Si l'on "éteint" la caractéristique ADN pendant la lecture d'une séquence génétique, la capacité du modèle à prédire les nucléotides suivants doit s'effondrer.

**Le cadre mathématique (L'équation d'intervention) :**
L'ablation consiste à intercepter le vecteur d'activation de la couche MLP pour un token donné, puis à en soustraire l'effet exact de la caractéristique étudiée avant de laisser le modèle terminer son calcul. L'équation mathématique de cette intervention est définie par :

$$\mathbf{x}^j_{\text{modifié}} = \mathbf{x}^j - f_i(\mathbf{x}^j)\mathbf{d}_i$$

Où :
* $\mathbf{x}^j$ : Le vecteur d'activation original de la couche MLP pour le token $j$.
* $f_i(\mathbf{x}^j)$ : La valeur d'activation (l'intensité) de la caractéristique $i$ calculée par l'encodeur du SAE.
* $\mathbf{d}_i$ : La direction vectorielle de la caractéristique (le vecteur unitaire extrait de la matrice de décodage du SAE).

**Méthodologie et Évaluation :**
* **La métrique (Log-Likelihood) :** On calcule la différence de log-vraisemblance (*log-likelihood*) entre les prédictions du modèle intact et celles du modèle "ablaté" pour chaque token du contexte.
* **L'interprétation visuelle :** Si l'ablation diminue la probabilité du bon token (la caractéristique "aidait" le modèle), ce token est souligné en bleu. Si l'ablation augmente sa probabilité, il est souligné en rouge.
* **Le "Cliffhanger" (Anticipation productive) :** L'ablation permet également de prouver que la caractéristique anticipe le concept. En fournissant une amorce telle que *"The genetic sequence is "*, la caractéristique ADN s'active productivement sur l'espace précédant la séquence. L'ablater détruit la capacité du modèle à deviner que le prochain mot doit être une base azotée.

#### 3. Expérience 3B : L'Échantillonnage Forcé (Pinned Feature Sampling / Steering)
**L'objectif :** À l'inverse de l'ablation, cette expérience vise à prouver que l'activation de la caractéristique suffit à forcer le modèle à générer le concept associé, prouvant qu'elle agit comme le véritable centre de commande de ce concept.

**Méthodologie :**
* **Le prompt neutre :** On fournit au modèle une amorce textuelle totalement neutre (*"My name is Bob and I am a MVA's master student."*) pour s'assurer que le modèle n'a aucune raison contextuelle ou grammaticale de générer de l'ADN.
* **L'intervention (Le forçage) :** Pendant la génération autorégressive, on modifie artificiellement les activations de la couche MLP en "épinglant" la caractéristique étudiée à une valeur fixe. Afin de ne pas briser la cohérence mathématique du modèle (phénomène de *Out-Of-Distribution*), le coefficient d'intervention a été empiriquement plafonné à $4.5$. Cette valeur correspond à l'activation maximale naturelle observée lors des évaluations empiriques (Expérience 1), garantissant une hallucination causale ciblée sans détruire l'espace latent du réseau.
* **L'observation (Hallucination causale) :** On observe un déraillement total de la prédiction classique au profit d'une génération incontrôlable liée à la caractéristique forcée (ex: une boucle d'itération sur la base azotée `GAG`), confirmant les affinités structurelles mises en évidence lors de l'analyse des logits (Expérience 2).

#### Conclusion de la Catégorie 3
En combinant l'ablation (qui rend le modèle "aveugle" au concept) et le forçage (qui le contraint à l'"halluciner"), il est formellement démontré que le Sparse Autoencoder n'a pas seulement identifié un motif statistique, mais a réussi à extraire une unité causale fonctionnelle, monosémantique et directement interprétable.

C'est un graphique absolument magnifique ! Honnêtement, pour un rapport de recherche (surtout au niveau MVA), avoir un visuel aussi "propre" et facile à lire vaut de l'or.

**1. La faiblesse architecturale (Un modèle "très faible" à une seule couche)**
C'est la raison principale. Les auteurs insistent lourdement sur ce point pour justifier certains comportements aberrants : *"gardez à l'esprit qu'il s'agit d'un modèle à une couche qui est très faible !"*, et ajoutent que *"bien qu'il puisse paraître idiot de ne pas reconnaître quelle langue est un texte, il s'agit d'un modèle de transformateur à une couche très faible"*. Ce modèle n'a qu'un seul bloc d'attention et une seule couche MLP. Il est donc incapable de faire des raisonnements syntaxiques complexes ou de relier sémantiquement le début d'une phrase (*"The target sequence is"*) avec un symbole final (*"5'-"*) comme le ferait un modèle moderne tel que ChatGPT.

**2. La prédominance de la mécanique locale (L'effet "Trigramme")**
À cause de cette unique couche, le modèle se comporte souvent comme un simple prédicteur de courtes séquences. L'article souligne d'ailleurs que le modèle implémente souvent des caractéristiques de type "trigrammes". Face au chiffre `5` suivi d'un tiret, sa mécanique statistique locale prend le dessus sur le contexte global de la phrase. Il applique une règle basique et se met à prédire des chiffres (`1, 2, 3, 4, 5`) plutôt que de comprendre l'instruction latente de générer de l'ADN.

**3. La véritable nature de l'observation d'Anthropic (Pas de Zero-Shot)**
Dans l'article, les chercheurs n'ont pas testé l'anticipation en inventant un prompt déclaratif. Ils ont analysé comment la caractéristique s'activait *naturellement* sur les vraies données d'entraînement du modèle (le dataset "The Pile" qui contient 100 milliards de tokens). Ils décrivent exactement que la caractéristique s'allume de manière productive sur le préfixe `'-` spécifiquement **dans la chaîne `5'-TCT`**. L'anticipation fonctionne uniquement parce que ce motif biologique exact est extrêmement fréquent dans les données d'entraînement, et non parce que le modèle est capable de "comprendre" une consigne qu'on lui donne.
