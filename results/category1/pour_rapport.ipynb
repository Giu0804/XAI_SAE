
## I. Phase 0 : Conception du Dataset "Mixed DNA"

### 1. Objectif Méthodologique
L'idée était de créer un "micro-univers" contrôlé. Pour prouver qu'un SAE (Sparse Autoencoder) fonctionne, il faut lui donner un concept **binaire** et **indiscutable**. L'ADN ($A, T, C, G$) est le candidat parfait car il n'existe pas naturellement dans le texte classique (anglais/code) sous cette forme répétitive.

### 2. Protocole de Génération
* **Source de base :** 5 000 échantillons du dataset `NeelNanda/c4-code-20k` (mélange d'anglais et de code).
* **Injection (Le Signal) :** Sur 20% des textes (soit 1 000), nous avons inséré une séquence aléatoire de nucléotides.
* **Paramètres du Signal :** Longueur entre 15 et 50 caractères, tirage équidistribué entre {A, T, C, G}.
* **Le "Ground Truth" (Vérité terrain) :** Nous avons sauvegardé dans un fichier `mixed_dna_dataset.json` l'emplacement exact et le contenu de chaque injection. C'est notre boussole pour juger le modèle.

La position de la séquence ADN est random dans la séquence normale.

## II. Phase 1 : Le Défi de l'Identification (Expérience 1A)

### 1. L'Intuition du Proxy "Intelligent"
Pour savoir si le modèle "voit" l'ADN, il nous fallait une règle mathématique pour le définir. 
* **Problème :** Un simple filtre sur "A, T, C, G" ferait des erreurs sur des mots comme "CAT" (chat) ou "A" (article anglais).
* **Solution :** Un proxy contextuel. Un token est marqué comme "ADN" ($1.0$) uniquement s'il est composé de $\{A, T, C, G\}$ **ET** qu'il est adjacent à un autre token lui ressemblant. Cela élimine les faux positifs linguistiques.

### 2. Le Match : SAE vs Neurone MLP
Nous avons comparé deux unités de calcul pour voir laquelle est la plus "intelligente" :
1.  **Le Neurone #1996 (Baseline) :** Un neurone "naturel" du modèle GPT.
2.  **La Caractéristique SAE #16094 :** Une unité reconstruite par le Sparse Autoencoder.

### 3. La Rigueur Mathématique (Corrélation de Pearson)
Nous avons mesuré la corrélation entre l'activation de ces unités et notre Proxy ADN. Pour éviter le crash de la RAM (dû à la matrice géante de 92 Go), nous avons utilisé la **formule incrémentale** :

$$r = \frac{n(\sum xy) - (\sum x)(\sum y)}{\sqrt{[n\sum x^2 - (\sum x)^2][n\sum y^2 - (\sum y)^2]}}$$

> **Résultats obtenus :**
> * **SAE Feature #16094 :** **0,7545** (Corrélation forte, le SAE a "isolé" le concept).
> * **Neuron #1996 :** **0,4701** (Corrélation faible, le neurone est "pollué").

### 1. La théorie du problème : La "Superposition" et le Polysémantisme
C'est la théorie qui explique le mauvais score de votre neurone de référence (la *Baseline*).
*   **L'explication :** Les auteurs expliquent que l'unité de base d'un réseau (le neurone) n'est pas naturelle pour la compréhension humaine car de nombreux neurones sont "polysémantiques", c'est-à-dire qu'ils réagissent à un mélange d'entrées qui n'ont apparemment aucun rapport. 
*   **La cause :** Ce phénomène est dû à la **"superposition"**, une hypothèse selon laquelle les réseaux de neurones doivent représenter plus de caractéristiques indépendantes qu'ils ne possèdent de neurones. Pour y parvenir, le modèle compresse les données en attribuant à chaque concept une combinaison linéaire de neurones. 

### 2. La théorie de la solution : La Décomposition Mathématique (Le SAE)
C'est la partie où vous expliquez mathématiquement comment votre autoencodeur creux "démêle" cette superposition.
*   **L'équation de décomposition :** Le but du SAE est de décomposer le vecteur d'activation original du modèle, noté $\mathbf{x}^j$, en une combinaison linéaire de caractéristiques plus générales. L'équation fondamentale du papier est la suivante : 
    **$\mathbf{x}^j \approx \mathbf{b} + \sum_i f_i(\mathbf{x}^j) \mathbf{d}_i$**.
    *Ici, $f_i(\mathbf{x}^j)$ est l'activation de la caractéristique $i$, $\mathbf{d}_i$ est un vecteur unitaire (la direction de la caractéristique), et $\mathbf{b}$ est un biais.*
*   **L'équation d'inférence (Votre code) :** Vous devez ensuite montrer comment on calcule concrètement l'activation de la caractéristique (le fameux $f_i(x)$). La formule mathématique exacte utilisée par l'encodeur est : 
    **$f_i(x) = \text{ReLU}( W_e (\mathbf{x} - \mathbf{b}_d) + \mathbf{b}_e )_i$**.
    *Vous pourrez faire remarquer à vos professeurs que c'est exactement cette équation qui est implémentée dans votre fonction `encode(self, x)` !*

### 3. La théorie de l'évaluation : Le ratio de Log-Vraisemblance (Le Proxy)
C'est la théorie qui justifie pourquoi vous avez codé une fonction de détection d'ADN et comment vous l'avez comparée aux activations.
*   **L'équation du Proxy :** Pour prouver qu'une caractéristique s'active spécifiquement pour un contexte donné (comme l'ADN), les auteurs définissent un proxy computationnel qui est une estimation du ratio de log-vraisemblance : 
    **$\log(P(s|\text{context}) / P(s))$**.
*   **L'application à l'ADN :** Théoriquement, l'ADN est modélisé comme une chaîne aléatoire constituée de caractères issus de `[ATCG]`, où chaque nucléotide a une probabilité de 1/4 d'apparaître. 
*   **La justification de votre code binaire :** Pourquoi avez-vous mis des scores de 1.0 et 0.0 au lieu de vraies probabilités logarithmiques complexes ? L'article vous donne l'argument théorique parfait : les auteurs affirment que puisque les probabilités logarithmiques s'avèrent être très bimodales, il est justifié de binariser le proxy. 
*   **La validation :** Enfin, la preuve ultime du monosémantisme est établie en calculant la corrélation de Pearson entre ce proxy binarisé et les activations réelles de la caractéristique.



## 1. La Problématique : La Superposition
Le point de départ de ton rapport doit être le suivant : les modèles de langage ont plus de concepts à apprendre que de neurones physiques. Pour compenser, ils utilisent la **superposition**.

L'idée théorique (proposée par Elhage et al.) est que le modèle représente les concepts comme des **directions** dans un espace de haute dimension, plutôt que par des neurones uniques. 
* **Neurone polysemantic :** Un neurone $n$ répond à plusieurs concepts non corrélés (ex: ADN + Recettes de cuisine).
* **Conséquence :** On ne peut pas comprendre le modèle en regardant juste ses neurones. C'est là qu'intervient le **SAE**.

---

## 2. L'Architecture du Sparse Autoencoder (SAE)
Ton SAE agit comme un "déplieur" de dimensions. Il prend une activation de couche cachée $x$ (ton MLP) et tente de la projeter dans un espace beaucoup plus grand mais très peu peuplé (sparse).

### L'Encodeur (La Détection)
C'est l'étape où le SAE cherche les caractéristiques (features) présentes. L'équation de l'activation de la feature $i$ est :

$$f_i(x) = \text{ReLU}((x - b_{pre})W_{enc, i} + b_{enc, i})$$

Où :
* $W_{enc, i}$ est le poids de l'encodeur pour la feature $i$.
* $b_{enc}$ est le biais qui force la "pauvreté" (sparsity) : il faut que le signal soit assez fort pour dépasser ce biais et activer la ReLU.

### Le Décodeur (La Reconstruction)
Le SAE doit être capable de reconstruire l'activation originale $x$ à partir de ses caractéristiques :

$$\hat{x} = b_{dec} + \sum_{i=1}^{M} f_i(x)W_{dec, i}$$

Le but du SAE est que $\hat{x} \approx x$ tout en gardant la majorité des $f_i(x)$ à zéro.

---

## 3. La Mesure de Fidélité : Corrélation de Pearson
Pour prouver que la feature $i$ est bien le "Concept ADN", tu as utilisé la corrélation de Pearson. Dans ton rapport, tu peux la formaliser ainsi :

Soit $X$ le vecteur des activations de la feature $i$ sur $N$ tokens, et $Y$ le vecteur des scores de ton Proxy ADN (0 ou 1). La corrélation $r$ est :

$$r = \frac{\text{cov}(X, Y)}{\sigma_X \sigma_Y}$$

Dans ton code, comme on avait un problème de mémoire (RAM), on a utilisé la **version incrémentale (Online Pearson)**. C'est un point technique très valorisant à mettre dans ton rapport :

> *"Pour traiter 750 000 tokens sans saturer la mémoire vive, nous avons implémenté une mise à jour itérative des moments d'ordre 1 et 2 (Sommes et Sommes des carrés), permettant de calculer le coefficient de corrélation sans stocker la matrice complète des activations."*

## 4. Définition de la Monosémanticité
C'est le cœur de ton Expérience 1B. Une feature est dite **monosémantique** si sa distribution d'activation est alignée avec un seul concept interprétable.

Théoriquement, cela se traduit par une **séparabilité linéaire** dans l'espace des activations. Tes histogrammes prouvent cette séparabilité :
* **Dans le MLP :** Les distributions "DNA" et "Not DNA" se chevauchent (Interférence).
* **Dans le SAE :** Les distributions sont disjointes pour les hautes activations (Pureté).
